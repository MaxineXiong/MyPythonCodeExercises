## 1. Math Equation Plotting

### Linear

$y = 2x + 1$

```python
plt.plot([1, 10], [3, 21], linestyle = '-.', 
                           linewidth = 2.0,
                           color = 'red', 
                           alpha = 0.5,     #alpha: transparency
                           marker = 'o'
        )  
```

#### see API doc for more styling options: https://matplotlib.org/3.5.0/api/_as_gen/matplotlib.pyplot.plot.html

```python
x = np.linspace(1, 10, 2)
y = 2* x + 1
```

```python
x
```

```python
y
```

```python
plt.plot(x, y, marker = '+')
```

### Logistic

$$y = \frac {1} {1 + e^{-z}}$$

```python
z = np.linspace(-10, 10, 100)

y = 1 / (1 + np.exp(-z))
plt.plot(z, y, ls = '-', c = 'red')
plt.show()
```

$$y = \frac {1} {1 + e^{-iz}}$$

```python
for i in [0.2, 0.5, 1, 2, 4]:
    y = 1 / (1 + np.exp(-i*z))
    plt.plot(z, y, '-', label=i)

plt.legend()
plt.show()
```

### Entropy

$$y = - p\log_2p - (1-p)\log_2(1-p)$$

```python
p = np.linspace(0.01, 1, 100, endpoint=False)
y = -p*np.log2(p) - (1-p)*np.log2(1-p)
plt.plot(p, y, '-')
plt.show()
```

## 2. Covid-19 Example

```python
import findspark
import os

findspark.init(os.environ['SPARK_HOME'])
```

```python
from pyspark.sql import SparkSession
import pyspark.sql.functions as F
```

```python
spark = SparkSession.builder.appName('visualisation').getOrCreate()
```

```python
df = spark.read.csv('../data/WHO-COVID-19-global-data.csv', header=True, sep = ',', inferSchema=True)
```

```python
print(df.count())
df.printSchema()
```

```python
df.show()
```

```python
df = df.withColumn('Date_reported', F.to_date('Date_reported')).cache()
```

```python
df.count()
```

```python
df.filter('Country = "China"').orderBy('Date_reported').show()
```

```python
r = (df.filter('Country = "China"')
     .orderBy('Date_reported')
     .select('Date_reported', 'Cumulative_cases', 'Cumulative_deaths')
     .collect())
```

## Line Chart

```python
x = [r.Date_reported for r in r]
y1 = [r.Cumulative_cases for r in r]
y2 = [r.Cumulative_deaths for r in r]
```

```python
plt.figure(figsize = (12, 5))

plt.plot(x, y1, label='Confirmed')
plt.plot(x, y2, label='Death')

plt.title('Covid-19 Confirmed v.s. Death in China')
plt.legend()
plt.xlabel('Date')
plt.tight_layout()
plt.show()
```

#### Trick in visualisation:
when the scale of y1 and y2 differs a lot, convert them to **log<sub>10</sub>** y1 and **log<sub>10</sub>** y2

```python
#fig = plt.figure(figsize=(12,4))

plt.plot(x, [np.log10(y) for y in y1], label='Confirmed (log10)')
plt.plot(x, [np.log10(y) for y in y2], label='Death (log10)')

plt.title('Covid-19 Confirmed v.s. Death in China')
plt.legend()
plt.xlabel('Date')
plt.ylabel('log10')
plt.tight_layout()
```

### Code Challenge

* 选5个国家
* 比较累计确诊人数
* 以第一天发现确诊病例为起始点
* 比较一个月内发展趋势的曲线
* 把指标换成确诊人数占总人口的比例

```python
from datetime import timedelta
import pandas as pd
```

```python
countries = ['China', 'Republic of Korea', 'Japan', 'Australia', 'United States of America', 'The United Kingdom']
```

```python
firstDates = {}
endDates = {}
```

```python
for c in countries:
    firstDate = df.filter(df.Country == c).where("Cumulative_cases > 0").sort('Date_reported').collect()[0].Date_reported
    firstDates[c] = firstDate
    endDates[c] = firstDate + timedelta(days = 30)
```

```python
firstDates
```

```python
endDates
```

```python
x = list(range(1, 31))

plt.figure(figsize = (10, 5))
for c in countries:
    y = [r.Cumulative_cases for r in df.filter(df.Country == c).filter( (df.Date_reported >= firstDates[c]) & (df.Date_reported < endDates[c]) ).collect()]
    plt.plot(x, y, label = c)
    
plt.xticks(x)
plt.legend(loc = 'upper left', bbox_to_anchor = (1.05, 1))
plt.xlabel('Day')
plt.ylabel('Confirmed Cases %')
plt.tight_layout()
plt.show()
```

## Bar Chart

```python
# 最新数据对应的日期
df.select(F.max('Date_reported')).show()
```

```python
# 截止到最新数据，累计死亡人数最多的前五名国家
df.filter('Date_reported = "2020-06-02"').orderBy('Cumulative_deaths', ascending=False).show(5)
```

```python
bar_data = (df.filter('Date_reported = "2020-06-02"')
            .orderBy('Cumulative_deaths', ascending=False)
            .select('Country_code', 'Cumulative_deaths')
            .head(5))
```

```python
codes = [r.Country_code for r in bar_data]
deaths = [r.Cumulative_deaths for r in bar_data]
```

```python
plt.bar(codes, deaths, align='center', alpha=0.8, color='#3366ff')
#plt.barh(codes, deaths, align='center', alpha=0.5)

plt.ylabel('Cumulative Deaths')
plt.xlabel('Country Code')
plt.title(f'Top {len(codes)} Countries as of 2020-06-02')
```

## Pie Chart

```python
plt.pie(deaths, labels=codes);
```

```python
plt.pie(deaths, labels=codes, startangle=90);
```

```python
plt.pie(deaths, labels=codes, startangle=90, autopct='%1.1f%%');
```

```python
explode = [0.1 if c=='BR' else 0.0 for c in codes]

plt.pie(deaths, labels=codes, startangle=90, autopct='%1.1f%%', explode=explode);

plt.title("A Pie Chart");
```

## Geo Map

[WHO Covid19 Dashboard](https://covid19.who.int/)

```python
cc = spark.read.csv('./data/countryMap.txt', header=True, inferSchema=True, sep='\t')
```

```python
cc = cc.select('2let', '3let').withColumnRenamed('2let', 'Country_code')
```

```python
cc.show()
```

```python
res = df.join(cc, 'Country_code')
```

```python
res.show()
```

```python
latest = (res.filter('Date_reported = "2020-06-02"')
          .select('3let', 'Country', 'Cumulative_cases', 'Cumulative_deaths')
          .toPandas())
```

```python
latest.head()
```

```python
import chart_studio.plotly as py
import plotly.graph_objs as go
from plotly.offline import iplot, plot
```

```python
data = dict(
        type = 'choropleth',
        locations = latest['3let'],
        z = latest['Cumulative_cases'],
        text = latest['Country'],
        colorbar = {'title' : 'Number of Confirmed Cases'},
      ) 
```

```python
layout = dict(
    title = 'Covid-19 Global Confirmed Cases on {}'.format('2020-06-02'),
    geo = dict(
        showframe = False
    )
)
```

```python
choromap = go.Figure(data = [data], layout = layout)
iplot(choromap)
```

```python
plot(choromap)
```
