```python
def PlotFunction(data, players = Players):
    plt.figure(figsize = (12, 6))
    for player in players:
        plt.plot(Seasons, data[Pdict[player]], label = player, 
                                               marker = Mdict[player], 
                                               color = Cdict[player],
                                               linestyle = '--',
                                               markersize = 8)   
    plt.legend(loc = 'upper left', bbox_to_anchor = (1, 1))
    plt.xticks(rotation = 45)
    plt.show()
```

```python
PlotFunction(Salary)
```

```python
PlotFunction(Salary, ["KobeBryant", 'LeBronJames', 'CarmeloAnthony'])
```

```python
PlotFunction(Games)
```

```python
PlotFunction(Salary)
PlotFunction(Salary / Games)
PlotFunction(Salary / FieldGoals)
```

Conclusion: 
* KobeBryant and DerrickRose attended fewer games and achieved hence fewer field goals
* Due to the fact that they got injured in 2013 but still recieved higher salary
* Players get paid regardless of whether or not they played games because they have contracts

```python
PlotFunction(FieldGoals / Games)
PlotFunction(FieldGoals / FieldGoalAttempts)
PlotFunction(FieldGoalAttempts / Games)
PlotFunction(Points / Games)
```

Conclusion: 
* KevinDurant threw the most field goals and scored the highest per game.
* CameloAnthony tried the most times to throw.
* DwightHoward had the highest accuracy
* The reason why DwightHoward didn't score the highest points is because he didn't have many chances to throw.

```python
PlotFunction(MinutesPlayed / Games)
PlotFunction(Games)
```

Conclusion:
* The games all players attended every year remains constant.
* Minutes they played in each game goes down over years.

```python
PlotFunction(FieldGoals / MinutesPlayed)
```

Conclusion:
* KevinDurant scored the most field goals per minute - the most efficient player.

```python
PlotFunction(Points / FieldGoals)
```

Conclusion:
* KevinDurant threw 3-point field goals the most
